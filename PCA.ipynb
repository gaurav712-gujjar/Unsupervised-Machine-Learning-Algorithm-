```python
# PCA is a dimensionality reduction technique in machine learning that transform a large set of correlated features into
# a smaller set of uncorrelated variables called principal components while preserving as much variance as possible

# Steps Of PCA
# (1)Standardize the data
# (2)Compute co-variance metrics of the feature
# (3)Calculate igon Eigenvector and Eigenvalues from the co variance metrics
# (4)Sort igon values in descending order
# (5)Select the top K igon vectors
# (6)Form the pricipam component metrics using selected igon vectors
# (7)Project the original data onto these principal components
# (8)Obtain reduced dimensional data set
```

Variance in PCA

In Principal Component Analysis, variance means:
.How much the data points are spread out.
.Higher variance = more information.



```python
import pandas as pd
import numpy as np
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
df.head(3)
```

```python
df = df.drop(columns=['User ID', 'Gender'])
```

```python
x = df.drop(columns=['Purchased'])
y = df['Purchased']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from numpy.random.mtrand import RandomState
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('model', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
