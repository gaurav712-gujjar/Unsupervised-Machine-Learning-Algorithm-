```python
# APYORI MODEL (Apriori Algorithm)
# The Apyori model is a Python implementation of the Apriori Algorithm, which is used in Association Rule Learning. It helps in finding relationships between items in a dataset.
# It is widely used in market basket analysis (e.g., which products are bought together).
# It works on the idea that:
# “If an itemset is frequent, then all of its subsets must also be frequent.”

# Working of Apriori:
# Generate all possible item combinations.
# Calculate their support.
# Remove itemsets with support less than minimum threshold.
# Generate association rules using confidence and lift.


# We use Apyori to implement the Apriori algorithm in Python,
# which helps in finding frequent itemsets and association rules between items in transactional data,
# mainly used in market basket analysis and recommendation systems.


# SUPPORT
# Support measures how frequently an itemset appears in the dataset.

# Support(A)=
# Total number of transactions
# Number of transactions containing A

# Value range: 0 to 1
# Higher support → item appears more often.

# 👉 Example:
# If 20 out of 100 transactions contain bread → support = 0.2





# CONFIDENCE
# Confidence measures how likely item B is purchased when item A is purchased.

# Confidence(A→B)=
# Support(A)
# Support(A∩B)

# Value range: 0 to 1
# Higher confidence → stronger rule.

# 👉 Example:
# If 10 transactions contain both A and B, and 20 contain A → confidence = 10/20 = 0.5





# LIFT
# Lift measures how much more likely item B is purchased with A compared to buying B alone.

# Lift(A→B)=
# Support(B)
# Confidence(A→B)

# Interpretation:
# Lift = 1 → No association (independent)
# Lift > 1 → Positive association (good rule)
# Lift < 1 → Negative association

# 👉 Example:
# If confidence = 0.5 and support(B) = 0.25 → lift = 2 (strong relation)
```

```python
'''
    Overall Workflow
    Step 1
    Load dataset

    Step 2
    Convert rows into transactions

    Step 3
    Apply Apriori algorithm

    Step 4
    Generate association rules using:
      . Support
      . Confidence
      . Lift

    Step 5
    Extract frequent itemsets
'''
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/market_basket - market_basket.csv", header=None)
df.head(3)
```

```python
# Data Processing
# convert rows into transaction
transactions = []
for i in range(0,7501):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])

#Training the dataset
from apyori import apriori
rules = apriori(transactions ,
    min_support = 0.003,
    min_confidence = 0.2,
    min_lift = 3,
    min_length = 2)

#Visualizing the results
a = list(rules)
result = [list(a[i][0]) for i in range(0 , len(a))]

#(3items * 7 days in week)/(Total Rows in dataset 7501)
```
